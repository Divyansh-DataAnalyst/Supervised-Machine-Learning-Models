```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.svm import SVC
```

```python
# feature :[cgpa,coding Skill]
x = np.array([[8.5,9],
              [8.0,8],
             [7.5,8],
             [6.0,5],
            [5.5,4],
              [6.2,5]
             ])
#  labels
y = np.array([1,1,1,0,0,0])
```

```python
# create SVM Model
model = SVC(kernel ='linear')
```

```python
# TRAIN THE MODEL
model.fit(x,y)
```

```python
new_student = np.array([[3.2,9]])
result = model.predict(new_student)

if result[0] == 1:
  print("Student will be placed")
else:
  print("Student will not be placed")
```

```python
plt.scatter(x[y==1][:,0],x[y==1][:,1],marker='o',label ='placed')
plt.scatter(x[y==0][:,0],x[y==0][:,1],marker="x",label="Not Placed")

w = model.coef_[0]
b = model.intercept_[0]

x_points = np.linspace(5,9)
y_points = -(w[0]*x_points + b) /w[1]

plt.plot(x_points,y_points,label ='hyperplne')

plt.xlabel('cgpa')
plt.ylabel("coding skill")
plt.legend()
plt.show()

```
